```python
# Importaciones necesarias para el notebook
import matplotlib.pyplot as plt
import numpy as np
import os
```

<h1>Line Plot</h1>

### 1. Line Plot: Tendencias de Temperatura

**¿Qué muestra el gráfico?**
Muestra la simulación de la variación de temperatura a lo largo de una semana para dos ciudades distintas (Ciudad A y Ciudad B). 

**¿Por qué elegí esta visualización?**
El gráfico de líneas es la visualización ideal para representar series temporales o datos continuos a lo largo del tiempo, ya que permite identificar rápidamente tendencias, picos y caídas en los datos.

```python
# Datos generados
dias = np.array(['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom'])
temp_ciudad_a = np.array([22, 24, 23, 26, 28, 27, 25])
temp_ciudad_b = np.array([18, 19, 18, 20, 22, 21, 19])

plt.figure(figsize=(10, 5))

# Creación del gráfico con estilos personalizados
plt.plot(dias, temp_ciudad_a, marker='o', linestyle='-', color='#FF5733', linewidth=2, label='Ciudad A')
plt.plot(dias, temp_ciudad_b, marker='s', linestyle='--', color='#3380FF', linewidth=2, label='Ciudad B')

# Etiquetas y Título
plt.title('Tendencia de Temperaturas Semanal', fontsize=14, fontweight='bold')
plt.xlabel('Días de la Semana', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)
plt.legend(loc='upper left')

#Guardar imagen
plt.savefig('imgs/tendencia-temperatura-semanal.png', dpi=300, bbox_inches='tight')
```

<h1>Bar Chart</h1>

### 2. Bar Chart: Comparación de Ventas

**¿Qué muestra el gráfico?**
Muestra los ingresos generados por las ventas de cuatro categorías diferentes de productos de tecnología durante el último trimestre.

**¿Por qué elegí esta visualización?**
El gráfico de barras es perfecto para comparar magnitudes entre diferentes categorías discretas. Facilita ver a simple vista qué categoría tuvo el mejor o peor rendimiento.

```python
# Datos generados
categorias = ['Laptops', 'Smartphones', 'Tablets', 'Monitores']
ventas = [45000, 62000, 28000, 35000]

plt.figure(figsize=(8, 6))

# Creación del gráfico con estilos (colores distintos y bordes)
colores = ['#4A90E2', '#50E3C2', '#F5A623', '#D0021B']
barras = plt.bar(categorias, ventas, color=colores, edgecolor='black', linewidth=1.5)

# Etiquetas y Título
plt.title('Ingresos por Ventas por Categoría (Q3)', fontsize=14, fontweight='bold')
plt.xlabel('Categoría de Producto', fontsize=12)
plt.ylabel('Ingresos Totales ($)', fontsize=12)

# Añadir leyenda simulada
plt.legend([barras], ['Ventas Q3'], loc='upper right')

#Guardar imagen
plt.savefig('imgs/comparacion-ventas.png', dpi=300, bbox_inches='tight')
```

<h1>Pie Chart</h1>

### 3. Pie Chart: Composición del Presupuesto

**¿Qué muestra el gráfico?**
Muestra la distribución porcentual del presupuesto anual de un departamento de IT, dividido en Hardware, Software, Personal y Mantenimiento.

**¿Por qué elegí esta visualización?**
El pie chart es ideal para mostrar la composición de un total (partes de un todo). Permite visualizar rápidamente qué categoría consume la mayor proporción de los recursos.

```python
# Datos para el gráfico de tarta
etiquetas = ['Hardware', 'Software', 'Personal', 'Mantenimiento']
proporciones = [25, 30, 35, 10]
colores_pie = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0, 0, 0.1, 0)  # "Explota" o separa la rodaja de 'Personal'

plt.figure(figsize=(8, 8))

# Creación del gráfico de tarta
plt.pie(proporciones, 
        explode=explode, 
        labels=etiquetas, 
        colors=colores_pie, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=140,
        textprops={'fontsize': 12})

# Título y Leyenda
plt.title('Distribución del Presupuesto Anual IT', fontsize=14, fontweight='bold')
plt.legend(etiquetas, title="Categorías", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

#Guardar imagen
plt.savefig('imgs/distribucion-presupuesto.png', dpi=300, bbox_inches='tight')
```

<h1>Histogram</h1>

### 4. Histogram: Distribución de Calificaciones

**¿Qué muestra el gráfico?**
Muestra la distribución de frecuencia de las calificaciones de un examen simulado para 500 estudiantes, siguiendo una distribución aproximadamente normal.

**¿Por qué elegí esta visualización?**
Los histogramas son fundamentales para entender cómo se distribuyen los datos en un conjunto unidimensional. Muestra dónde se concentran los valores, los rangos comunes y la presencia de asimetrías.

```python
# Datos generados aleatoriamente con NumPy (Distribución normal)
np.random.seed(42)
calificaciones = np.random.normal(loc=75, scale=10, size=500)
# Limitar a un máximo de 100
calificaciones = np.clip(calificaciones, 0, 100)

plt.figure(figsize=(9, 5))

# Creación del histograma
n, bins, patches = plt.hist(calificaciones, bins=20, color='#9B59B6', edgecolor='white', alpha=0.8, label='Frecuencia de Notas')

# Etiquetas y Título
plt.title('Distribución de Calificaciones del Examen Final', fontsize=14, fontweight='bold')
plt.xlabel('Calificación (0 - 100)', fontsize=12)
plt.ylabel('Cantidad de Estudiantes', fontsize=12)
plt.legend()

#Guardar imagen
plt.savefig('imgs/distribucion-calificaciones.png', dpi=300, bbox_inches='tight')
```

<h1>Scatter Plot</h1>

### 5. Scatter Plot: Relación Horas de Estudio vs. Calificación

**¿Qué muestra el gráfico?**
Representa la relación entre la cantidad de horas que un grupo de estudiantes dedicó al estudio y la nota final que obtuvieron en un examen.

**¿Por qué elegí esta visualización?**
El Scatter plot es la herramienta perfecta para visualizar correlaciones. En este caso, permite observar si a mayores horas de estudio, las calificaciones tienden a subir, ayudando a identificar patrones visuales de manera directa.

```python
# Datos generados: Horas de estudio vs Notas
np.random.seed(20)
horas_estudio = np.random.uniform(1, 20, 50)  # 50 estudiantes de 1 a 20 horas
notas = horas_estudio * 4 + np.random.normal(0, 10, 50)  # Nota con algo de ruido aleatorio
notas = np.clip(notas, 0, 100) # Asegurar que las notas estén entre 0 y 100

plt.figure(figsize=(9, 6))

# Creación del gráfico de dispersión
plt.scatter(horas_estudio, notas, 
            color='teal', 
            marker='X', 
            s=100, 
            alpha=0.7, 
            edgecolors='black', 
            label='Estudiantes')

# Título y etiquetas de los ejes con unidades
plt.title('Relación: Esfuerzo de Estudio y Rendimiento', fontsize=14, fontweight='bold')
plt.xlabel('Horas de Estudio Semanales', fontsize=12)
plt.ylabel('Calificación Obtenida (Puntos)', fontsize=12)

# Añadir una cuadrícula y leyenda[cite: 1]
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

#Guardar imagen
plt.savefig('imgs/relacion-estudio-calificaciones.png', dpi=300, bbox_inches='tight')
```
